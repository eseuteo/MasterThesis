```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sepsis3-df.csv')
```

```python
df[df.excluded == 0]
```

```python
times_parallelism_1 = [630735.5, 332719.5, 181623.5, 95711.5, 47588, 27196.5]
times_parallelism_2 = [433355,208254,136176,57958,50285,28232]
times_parallelism_4 = [288441,122782,95262,40136,31101,27696]
times_parallelism_8 = [251067,102340,77697,41601,31736,28509]
```

```python
times_parallelism_1.reverse()
times_parallelism_2.reverse()
times_parallelism_4.reverse()
times_parallelism_8.reverse()
```

```python
df = pd.DataFrame({'Sequential': times_parallelism_1, "Parallel (2)": times_parallelism_2, "Parallel (4)": times_parallelism_4, "Parallel (8)": times_parallelism_8, "Number of patients" : [1, 2, 4, 8, 16, 32]})
```

```python
plt.figure(figsize=(10, 5), dpi=80)
plt.plot("Number of patients", "Sequential", marker='o', data=df)
plt.plot("Number of patients", "Parallel (2)", marker='o', data=df)
plt.plot("Number of patients", "Parallel (4)", marker='o', data=df)
plt.plot("Number of patients", "Parallel (8)", marker='o', data=df)
plt.xticks(df["Number of patients"])
plt.ylabel("Execution time (ms)")
plt.xlabel("Number of patients")
plt.legend()
```
